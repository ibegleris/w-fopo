```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special

j0,j1,k0,k1 = scipy.special.j0,scipy.special.j1,scipy.special.k0,scipy.special.k1
```

```python
x = np.linspace(0,20,10000)
```

```python
for l,i in enumerate((j0,j1)):
    ii = i(x)
    plt.plot(x,ii,label = str(l))
  
plt.legend()
plt.show()
```

```python
x = np.linspace(0,5,10000)
for l,i in enumerate((k0,k1)):
    ii = i(x)
    plt.plot(x,ii,label = str(l))
plt.xlim(0,5)
plt.ylim(0,10)
plt.legend()
plt.show()
```

```python
x = np.linspace(0,0.0000001,10000)
kv = scipy.special.kv

plt.plot(x,kv([1],x),label = '0')
plt.xlim(0,5)
plt.ylim(0,10)
plt.legend()
plt.show()
```
